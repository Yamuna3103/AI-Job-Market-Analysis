VISUALIZATION

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure()
final_df['job_title'].value_counts().plot(kind='bar')
plt.title("Job Role Distribution")
plt.xlabel("Job Role")
plt.ylabel("Count")
plt.show()
```

```python
#Top 10 Most In-Demand Skills
plt.figure()
final_df['skills_required'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Most In-Demand Skills")
plt.xlabel("Skills")
plt.ylabel("Count")
plt.show()
```

```python
#Skills Distribution (Histogram-style)
plt.figure(figsize=(10,6))
sns.countplot(
    y=final_df['skills_required'],
    order=final_df['skills_required'].value_counts().head(15).index
)
plt.title("Top 15 Skills Distribution")
plt.xlabel("Count")
plt.ylabel("Skills")
plt.show()
```

```python
#Top Companies Hiring by Skills
plt.figure()
final_df['company_name'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Companies by Skill Demand")
plt.xlabel("Company Name")
plt.ylabel("Number of Skills Listed")
plt.show()
```

```python
#Company vs Skill Heatmap
pivot_table = pd.crosstab(
   final_df['company_name'],
    final_df['skills_required']
)

top_companies = pivot_table.head(10)
top_skills = top_companies.iloc[:, :10]

plt.figure()
sns.heatmap(top_skills,annot=True)
plt.title("Top Companies vs Top Skills Heatmap")
plt.show()
```

```python
#Horizontal Bar Chart
plt.figure()
final_df['skills_required'].value_counts().head(10).plot(kind='barh')
plt.title("Top 10 Skills (Horizontal)")
plt.xlabel("Count")
plt.ylabel("Skills")
plt.show()
```

```python
#Pie Chart
plt.figure()
final_df['skills_required'].value_counts().head(5).plot(kind='pie', autopct='%1.1f%%')
plt.title("Top 5 Skills Share")
plt.ylabel("")
plt.show()
```
